# The Power of Language: An Innovative Approach to Forecasting Financial Results

### Data Collection: ###
Use the Bloomberg API to fetch historical price data for the company's equity and debt issues.
Also, fetch news data. Bloomberg API provides news relevant to specific tickers, and you can use timestamps to understand when the news came out.

### Pre-processing: ### 
Clean the data to remove any missing values or outliers.
Convert news data into a binary format where 1 indicates the presence of news and 0 indicates no news.

### Event Study Methodology: ###
Define an event window around the news release (e.g., 5 minutes before and 30 minutes after).
Compute the average returns for equity and debt during non-event periods. These are called "normal" returns.
Subtract these normal returns from the returns during the event window. The difference gives you the abnormal returns which capture the effect of the news.

### Statistical Test: ###
You can use a t-test to see if the abnormal returns are statistically different from zero. A significant result indicates the news had an impact.
Cumulative Abnormal Returns (CAR):
Sum the abnormal returns over the event window to get the cumulative effect.

### Lag Analysis: ###
To understand the speed of adjustment, analyze how quickly the abnormal returns converge to zero after the news. The quicker they converge, the faster the market has assimilated the news.

### Multivariate Analysis (Optional): ###
If you want to control for other factors (like market returns), you can use a regression framework. For instance, you might regress the company's returns on market returns and a news dummy variable.

### Visualization: ###
Plot the cumulative abnormal returns over the event window. Peaks and troughs will provide insights into when the maximum impact was felt.

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']

import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp
from xbbg import blp
import seaborn as sns


# Data fetching
def fetch_data(equity_ticker, debt_ticker, start, end):
    equity_data = blp.bdh(equity_ticker, ['PX_LAST'], start_date=start, end_date=end)
    debt_data = blp.bdh(debt_ticker, ['PX_LAST'], start_date=start, end_date=end)
    
    equity_news_data = blp.bdh(equity_ticker, ['NEWS_HEAT_PUB_DAVG'], start_date=start, end_date=end)
    
    return equity_data, debt_data, equity_news_data

# Pre-processing
def compute_returns(data, ticker):
    data['returns'] = data[ticker].pct_change().dropna()
    return data

# Event Study
def abnormal_returns(data, news_data, equity_ticker):
    mask = news_data[(equity_ticker, 'NEWS_HEAT_PUB_DAVG')] == 0
    mask_reindexed = mask.reindex(data.index, fill_value=False)
    normal_returns = data.loc[mask_reindexed, 'returns'].mean()
    abnormal = data['returns'] - normal_returns
    return abnormal

# Statistical Test
def test_significance(abnormal_returns):
    t_stat, p_val = ttest_1samp(abnormal_returns.dropna(), 0)
    return t_stat, p_val

# Cumulative Abnormal Returns
def compute_CAR(abnormal_returns):
    return abnormal_returns.cumsum()

# Visualization
def visualize_CAR(equity_data, debt_data):
    plt.figure(figsize=(12, 6))
    plt.plot(equity_data.index, equity_data['CAR'], label='Equity CAR')
    plt.plot(debt_data.index, debt_data['CAR'], label='Debt CAR', linestyle='--')
    plt.legend()
    plt.title('Cumulative Abnormal Returns over Time')
    plt.xlabel('Date')
    plt.ylabel('CAR')
    plt.show()

# KDE visualization
def visualize_kde(equity_data, debt_data):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(equity_data['returns'], label='Equity', shade=True, color="b")
    sns.kdeplot(debt_data['returns'], label='Debt', shade=True, color="r")
    plt.title("Distribution of Returns: Equity vs. Debt")
    plt.xlabel("Returns")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()
    
def visualize_CAR_over_time(equity_data, debt_data):
    plt.figure(figsize=(12, 6))
    plt.plot(equity_data.index, equity_data['CAR'], label='Equity CAR')
    plt.plot(debt_data.index, debt_data['CAR'], label='Debt CAR', linestyle='--')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='dashed', label='Zero CAR')
    plt.legend()
    plt.title('Cumulative Abnormal Returns over Time')
    plt.xlabel('Date')
    plt.ylabel('CAR')
    plt.show()

# def run_analysis(equity_ticker, debt_ticker, start, end):
#     equity_data, debt_data, equity_news_data = fetch_data(equity_ticker, debt_ticker, start, end)
    
#     equity_data = compute_returns(equity_data, equity_ticker)
#     debt_data = compute_returns(debt_data, debt_ticker)
    
#     equity_abnormal_returns = abnormal_returns(equity_data, equity_news_data, equity_ticker)
#     debt_abnormal_returns = abnormal_returns(debt_data, equity_news_data, equity_ticker)
    
#     equity_t_stat, equity_p_val = test_significance(equity_abnormal_returns)
#     debt_t_stat, debt_p_val = test_significance(debt_abnormal_returns)
    
#     equity_data['CAR'] = compute_CAR(equity_abnormal_returns)
#     debt_data['CAR'] = compute_CAR(debt_abnormal_returns)
    
#     visualize_CAR_over_time(equity_data, debt_data)
#     visualize_CAR(equity_data, debt_data)
#     visualize_kde(equity_data, debt_data)
    
#     # Convert index to datetime for equity_news_data
#     equity_news_data.index = pd.to_datetime(equity_news_data.index)
    
#     # Calculate lag for equity
#     equity_peak_date = equity_data['CAR'].idxmax()
#     equity_lags = (equity_peak_date - equity_news_data.index).days
#     equity_avg_lag = equity_lags.mean()
    
#     # Calculate lag for debt
#     debt_peak_date = debt_data['CAR'].idxmax()
#     debt_lags = (debt_peak_date - equity_news_data.index).days
#     debt_avg_lag = debt_lags.mean()

#     print("Equity t-stat:", equity_t_stat)
#     print("Equity p-val:", equity_p_val)
#     print("Equity average lag (days):", equity_avg_lag)
    
#     print("Debt t-stat:", debt_t_stat)
#     print("Debt p-val:", debt_p_val)
#     print("Debt average lag (days):", debt_avg_lag)    # Calculate lag for equity
#     equity_peak_date = equity_data['CAR'].idxmax()
#     equity_lags = (equity_peak_date - equity_news_data.index.to_series().dt.to_timestamp()).dt.days
#     equity_avg_lag = equity_lags.mean()
    
#     # Calculate lag for debt
#     debt_peak_date = debt_data['CAR'].idxmax()
#     debt_lags = (debt_peak_date - equity_news_data.index.to_series().dt.to_timestamp()).dt.days
#     debt_avg_lag = debt_lags.mean()

#     print("Equity t-stat:", equity_t_stat)
#     print("Equity p-val:", equity_p_val)
#     print("Equity average lag (days):", equity_avg_lag)
    
#     print("Debt t-stat:", debt_t_stat)
#     print("Debt p-val:", debt_p_val)
#     print("Debt average lag (days):", debt_avg_lag)

def run_analysis(equity_ticker, debt_ticker, start, end):
    equity_data, debt_data, equity_news_data = fetch_data(equity_ticker, debt_ticker, start, end)
    
    equity_data = compute_returns(equity_data, equity_ticker)
    debt_data = compute_returns(debt_data, debt_ticker)
    
    equity_abnormal_returns = abnormal_returns(equity_data, equity_news_data, equity_ticker)
    debt_abnormal_returns = abnormal_returns(debt_data, equity_news_data, equity_ticker)
    
    equity_t_stat, equity_p_val = test_significance(equity_abnormal_returns)
    debt_t_stat, debt_p_val = test_significance(debt_abnormal_returns)
    
    equity_data['CAR'] = compute_CAR(equity_abnormal_returns)
    debt_data['CAR'] = compute_CAR(debt_abnormal_returns)

    visualize_CAR(equity_data, debt_data)
    
    # Convert index to datetime for equity_news_data
    equity_news_data.index = pd.to_datetime(equity_news_data.index)
    
    # Convert equity_peak_date to datetime
    equity_peak_date = pd.to_datetime(equity_data['CAR'].idxmax())
    
    # Calculate lag for equity
    equity_lags = (equity_peak_date - equity_news_data.index).days
    equity_avg_lag = equity_lags.mean()
    
    # Convert debt_peak_date to datetime
    debt_peak_date = pd.to_datetime(debt_data['CAR'].idxmax())
    
    # Calculate lag for debt
    debt_lags = (debt_peak_date - equity_news_data.index).days
    debt_avg_lag = debt_lags.mean()

    print("Equity t-stat:", equity_t_stat)
    print("Equity p-val:", equity_p_val)
    print("Equity average lag (days):", equity_avg_lag)
    
    print("Debt t-stat:", debt_t_stat)
    print("Debt p-val:", debt_p_val)
    print("Debt average lag (days):", debt_avg_lag)


start = '2017-01-01'
end = '2023-01-01'
equity_ticker = 'AAPL EQUITY'
debt_ticker = 'BBG00C7QBG91 Corp'

run_analysis(equity_ticker, debt_ticker, start, end)
```
